## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Identify top 5 spenders by total purchase value (SN,Purchase Count, Avg Purchase Price, Total Purchase Price)
#identify unique names through SN, group all values by SN
big_spenders_grouped = purchase_data.groupby("SN")

# similar to above use P ID count to get number of purchases
spender_purchase_count = big_spenders_grouped["Purchase ID"].count()

#total Purchase Price per person
spender_total_price = big_spenders_grouped["Price"].sum()

#Average Purchase Price
spender_avg_purch = spender_total_price/spender_purchase_count

big_spenders_df = pd.DataFrame({"Purchase Count": spender_purchase_count,
                                "Average Purchase Price": spender_avg_purch,
                                "Total Purchase Value": spender_total_price   
})
#big_spenders_df.head()
# Time to sort in descending order, top 5 values
big_spenders_df = big_spenders_df.sort_values("Total Purchase Value", ascending=False)

# dollar formatting for Avg. Purchase price  similar to pandas mapping
big_spenders_df["Average Purchase Price"] = big_spenders_df["Average Purchase Price"].map("${:.2f}".format)
# dollar formatting for Total Purchase Value similar to pandas mapping
big_spenders_df["Total Purchase Value"] = big_spenders_df["Total Purchase Value"].map("${:.2f}".format)

#Take top 5 values
big_spenders_df.iloc[0:5,:]
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Need to display 2 grouped identifiers
mp_items_grouped=purchase_data_df.groupby(["Item ID","Item Name"])

#use grouped workspace to calculate the amount (purchase count) and price, and total purchase value

#amount of purchases
mp_amount = mp_items_grouped["Purchase ID"].count()
#All prices for items are the same, just need a singular value
item_price = mp_items_grouped["Price"].mean()
#total purchase value is sum of all transactions for that item
total_item_value = mp_items_grouped["Price"].sum()

#Seriously capitalize the damn F
most_pop_items_df = pd.DataFrame({"Purchase Count": mp_amount,
                                  "Item Price": item_price,
                                  "Total Purchase Value": total_item_value   
})

# Time to sort in descending order, top 5 values
most_pop_items_df = most_pop_items_df.sort_values("Purchase Count", ascending=False)

# dollar formatting for Purchase value and price similar to pandas mapping
most_pop_items_df["Item Price"] = most_pop_items_df["Item Price"].map("${:.2f}".format)
#Total Purchase value
most_pop_items_df["Total Purchase Value"] = most_pop_items_df["Total Purchase Value"].map("${:.2f}".format)

#Take top 5 values
most_pop_items_df.iloc[0:5,:]
```
